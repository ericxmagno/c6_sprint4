```python
import pandas as pd
import numpy as np
import glob
```

```python
first = pd.read_csv('final/movies_2010.csv') 
```

```python
files = glob.glob('final/*.csv')

df_all = first.copy()

for file in files:
    if 'movies_2010' not in file:
        file_df = pd.read_csv(file)
        df_all= pd.concat([df_all, file_df], axis = 0, sort=False)
```

```python
df_all
```

```python
#just to check
years = np.arange(2010,2021)
```

```python
for year in years:
    print(year)
    print(df_all[df_all['Year']==year].count())
    print()
```

```python
df_all.to_csv('movies_all.csv', index=False)
```
