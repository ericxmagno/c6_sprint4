```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
```

```python
base_url = "https://en.wikipedia.org/wiki/List_of_Philippine_films_of_the_2000s"
```

```python
src = requests.get(base_url, headers = {'User-Agent':'Mozilla/5.0 (X11; Linux)'})
```

```python
soup = BeautifulSoup(src.content, 'lxml')
```

```python
tab_tag = soup.find_all('table', {'class':'wikitable'})
len(tab_tag)
```

```python
lst = []
```

```python
df = pd.DataFrame()
```

```python
count = np.arange(10)
print(count)
```

```python
for i in count:
    tab_ = pd.read_html(str(tab_tag[i]))
    df_ = pd.DataFrame(tab_[0])
    df = pd.concat([df, df_], axis = 0)
```

```python
movie_title = [x for x in df['Title']]
```

```python
df.reset_index(inplace=True)
df
```

```python
years = np.arange(2000,2010)
```

```python
years
```

```python
df.Year = 0
```

```python
# create year column
for year in years:
    if year == 2009:
        nidx = df.iloc[-1].name + 1
    else:
        nidx = df.loc[df.Title == str(year+1)].index[0]
    idx = df.loc[df.Title == str(year)].index[0]
    df.loc[idx:nidx, 'Year'] = year
```

```python
df = df[['Title', 'Director', 'Year']]
df.dropna(subset=['Title'], inplace=True)
df.drop_duplicates(subset=['Title'], inplace=True)
```

```python
df
```

```python
# drop year rows
for year in years:
    idx = df.loc[df.Title == str(year)].index[0]
    df.drop([idx], inplace=True)
df.reset_index(inplace=True, drop=True)
df
```

```python
df.to_csv('movies_2000s.csv', index=False)
```

```python
years = np.arange(2010, 2012)
```

```python
base_url = "https://en.wikipedia.org/wiki/List_of_Philippine_films_of_"

# 2010, 2011
for year in years:
    src = requests.get(base_url + str(year), headers = {'User-Agent':'Mozilla/5.0 (X11; Linux)'})
    soup = BeautifulSoup(src.content, 'lxml')
    tab_ten = soup.find_all('table', {'class':'wikitable'})
    df_ = pd.read_html(str(tab_ten))
    if year == 2010:
        df_ten = pd.DataFrame(df_[0])
    elif year == 2011:
        df_ten = pd.DataFrame(df_[1])
    df_ten.dropna(subset=['Title'], inplace=True)
    df_ten.drop_duplicates(subset=['Title'], inplace=True)
    df_ten = df_ten[['Title', 'Director']]
    df_ten['Year'] = year
    df_ten.to_csv(f'movies_{str(year)}.csv', index=False)
```

```python
# 2012
base_twl_url = "https://en.wikipedia.org/wiki/List_of_Philippine_films_of_2012"
src = requests.get(base_twl_url, headers = {'User-Agent':'Mozilla/5.0 (X11; Linux)'})
soup = BeautifulSoup(src.content, 'lxml')
tab_twl = soup.find_all('table', {'class':'wikitable'})
count_ = np.arange(1,5)
df_merged = pd.DataFrame()
for i in count_:
    tab_ = pd.read_html(str(tab_twl[i]))
    df_ = pd.DataFrame(tab_[0])
    df_merged = pd.concat([df_merged, df_], axis = 0)

df_merged.dropna(subset=['Title'], inplace=True)
df_merged.drop_duplicates(subset=['Title'], inplace=True)

df_merged = df_merged[['Title', 'Production company']]
df_merged.rename(columns={"Production company": "Director"}, inplace=True)
df_merged['Year'] = 2012
df_merged.to_csv('movies_2012.csv', index=False)

df_merged
```

```python
from urllib.request import urlopen

base_url = "https://en.wikipedia.org/wiki/List_of_Philippine_films_of_2013"

# 2013
page = urlopen(base_url)
page_content = page.read()
with open('2013.html', 'wb') as fid:
    fid.write(page_content)
```

```python
with open('2013.html', 'rb') as html:
    soup = BeautifulSoup(html, 'html.parser')
    tab_trt = soup.find_all('table', {'class':'wikitable'})
    count_ = np.arange(1,5)
    df_merged = pd.DataFrame()
    for i in count_:
        tab_ = pd.read_html(str(tab_trt[i]))
        df_ = pd.DataFrame(tab_[0])
        df_merged = pd.concat([df_merged, df_], axis = 0)
    df_merged.dropna(subset=['Title'], inplace=True)
    df_merged.drop_duplicates(subset=['Title'], inplace=True)
    df_merged = df_merged[df_merged['Cast and crew'].str.contains('Director')]
    df_merged['Director'] = df_merged['Cast and crew'].map(lambda s: s.split(' (Director')[0])
    df_merged = df_merged[['Title', 'Director']]
    df_merged['Year'] = 2013
    df_merged.to_csv(f'movies_2013.csv', index=False)
```
