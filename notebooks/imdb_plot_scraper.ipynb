```python
import pandas as pd
import numpy as np
import imdb
```

```python
df_movies = pd.read_csv('movies_2013.csv')
```

```python
df_movies
```

```python
import imdb
from tqdm import tqdm

ia = imdb.IMDb()

bar = tqdm(df_movies.iterrows())

for index, row in bar:
    movies = ia.search_movie(row['Title'])
    if movies:
        correct_movie = None
        for movie in movies:
            bar.set_description(row['Title'])
            m = ia.get_movie(movie.getID())
            if m.get('year') == row['Year']:
                correct_movie = m
            if correct_movie:
                break
        if correct_movie:
            plot = m.get('plot')
            if plot:
                df_movies.loc[index, 'Plot'] = plot[0]
            genre = m.get('genre')    
            if genre:
                df_movies.loc[index, 'Genre'] = ', '.join(genre)
```

```python
# update csv with plot 
df_movies.to_csv('movies_2013.csv', index=False)
```
